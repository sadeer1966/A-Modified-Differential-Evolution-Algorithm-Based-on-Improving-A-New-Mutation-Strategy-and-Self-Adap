```python
# عمل دالة تقوم بإرجاع قيمة الفهرسة
def return_index(lst, element):
    ind = 0
    index = []
    for i in lst:
        if i == element:
            index.append(ind)
        ind += 1
    return index
```

```python
with open('test_func.csv', 'r') as fl:
    contents = fl.readlines()[1:]
DOE_table = pd.read_csv('DOE_2.csv')
# eval_lst = []
results = pd.DataFrame()
# Loop over the required functions   
for j in contents:   
    # The output list is to save the output of applying DE using DOE trail on the current j function
    output = []
    for i in range(len(DOE_table)):
        trail = list(DOE_table.iloc[i, :])
        pop_size = trail[0]
        iteration_num = trail[1] 
        sol = DEalgorithm(pop_size, j[:-1], 5, iteration_num)
        # Equation 8 in the paper
        # output.append(0.6 * sol[0] + 0.4 * sol[2])
        output.append(sol[0] - 1/ (1+ sol[2]))
        
    results[j[:-1]] = output
```
